```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = sns.load_dataset("tips") ###load dataset from seaborn
```

```python
plt.title("made by seaborn")
sns.histplot(data["total_bill"])
plt.show()
```
