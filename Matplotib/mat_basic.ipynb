```python
value= np.array([78,21,1])
env = (["nitrogen","oxygen","carbon-dioxide"])
c =np.array(["blue","orange","black"])
plt.pie(value,labels =env, colors=c,
autopct='%1.1f%%',
textprops={'color':"green", 'fontsize':15},
shadow=True)
plt.title("Atmospheric Gas Composition")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

value = np.array([78, 21, 1])
env = ["nitrogen", "oxygen", "carbon-dioxide"]
colors = ["blue", "orange", "black"]

# Create pie and get the label handles
wedges, texts, autotexts = plt.pie(
    value,
    labels=env,
    colors=colors,
    autopct='%1.1f%%',
    shadow=True,
    textprops={'fontsize': 15}  # Only size here
)

# Set different colors for autopct text
autopct_colors = ['white', 'black', 'yellow']

for i, autotext in enumerate(autotexts):
    autotext.set_color(autopct_colors[i])
    autotext.set_fontweight('bold')

plt.title("Atmospheric Gas Composition")
plt.show()
```
